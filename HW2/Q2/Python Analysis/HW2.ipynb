```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Question 2.4: plot magnetization as a function of MC time for kJ/T = 2.26

# Open the file and read numbers into an array (list)
with open('M226.txt', 'r') as f:
    mag226 = [float(line.strip()) for line in f]  

# Generate an array from 0 to length of the numbers_array
length = len(mag226)
index_array = list(range(length))

# Plot the magnetization as a function of Monte Carlo time
fig, ax = plt.subplots()
plt.plot(index_array, mag226)
plt.xlabel('Monte Carlo Time')
plt.ylabel('Magnetization')
plt.grid(True)
plt.title('Magnetization vs. Monte Carlo Time (kT/J=2.26)')
plt.savefig('Magnetization226.png', dpi=300, bbox_inches='tight')
plt.show()
plt.clf()
```

```python
## Question 2.4: make a histogram of values for kJ/T = 2.26

# Plot a histogram of the magnetization values
fig, ax = plt.subplots()
plt.hist(mag226, bins=50)
plt.xlabel('Magnetization')
plt.ylabel('Counts')
plt.grid(True)
plt.title('Histogram of Magnetization Values (kT/J=2.26)')
plt.savefig('Histogram226.png', dpi=300, bbox_inches='tight')
plt.show()
plt.clf()
```
